# Activation Functions


```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def sigma(x):
    print ("the sigmoid Function", "\n", "\n", '1/(1+np.exp(-x))')
    return 1 / (1+np.exp(-x))
x=np.linspace(-5,5,100)
  #generates random numbers between -5 and 5

plt.plot(x,sigma(x),'b')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Sigmoid function')
plt.grid()
```

```python
def ReLu(x):
    return np.maximum(0.0,x)
print ("the ReLU Function", "\n", "\n", 'max(0,x)')
def ReLu_derivation(x):
    if x <=0:
        return 0
    else:
        return 1
    
plt.plot(x,ReLu(x),'b')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('ReLu Function')
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def sigmoid(x):
    return 1./(1+np.e**(-x))

# Calculate plot points
x = np.arange(-6., 6., 0.01)
y = sigmoid(x)
dx = y*(1-y)

# Setup centered axes
fig, ax = plt.subplots(figsize=(9, 5))
ax.spines['left'].set_position('center')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

# Create and show plot
ax.plot(x,y, color="#307EC7", linewidth=3, label="sigmoid")
ax.plot(x,dx, color="#9621E2", linewidth=3, label="derivative")
ax.legend(loc="upper right", frameon=False)
fig.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def softmax(x):
    #print ("softmax function", "\n", "\n", 'exp(x-max(x))')
    e_x = np.exp(x - np.max(x))
    return e_x /e_x.sum()

x=np.linspace(-5,5,100)
  #generates random numbers between -5 and 5

plt.plot(x,softmax(x),'b')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('softmax function')
plt.grid()
plt.show()
```
